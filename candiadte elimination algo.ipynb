```python
#correct classfy
import pandas as pd
import numpy as np
header=['shape','size','color','surface','thickness','target']
data=[['circular','large','light','smooth','thick','yes'],
     ['circular','large','light','irregular','thick','yes'],
     ['oval','large','dark','smooth','thin','no'],
     ['oval','large','light','irregular','thick','yes']]
data=pd.DataFrame(data,columns=header)
data.to_csv('forcast',index=False)
data=pd.read_csv('forcast')
concepts = np.array(data.iloc[:,0:-1])
target = np.array(data.iloc[:,-1])
l=[['circular','oval'],
   ['large'],
   ['light','dark'],
   ['smooth','irregular'],
   ['thick','thin']]
specific_h = ["0" for i in range(len(concepts[0]))]
print("\nInitialization of specific_h and genearal_h")
print("\nSpecific Boundary: ", specific_h)
general_h = [["?" for i in range(len(specific_h))] for i in range(len(specific_h))]
print("\nGeneric Boundary: ",general_h)
def check(y):
    j=0
    for i, h in enumerate(concepts):#elimination of dissatisfied hypothesis
        for k in range(len(specific_h)):

            if (general_h[k][k]==h[k] or general_h[k][k]=="?") and target[j]=="yes":
                continue
            if (general_h[k][k]!=h[k] and general_h[k][k]!="?") and target[j]=="no":
                continue
            else:
                general_h[k][k]="?"
        if y==j:
            break
        j=j+1
        

def k(val,col):
        for j in range(len(l[col])):
            if l[col][j]!=val:
                return l[col][j]
    
def learn(concepts, target):
      
    for i, h in enumerate(concepts):
        if target[i] == "yes":
            for x in range(len(specific_h)): 
                if h[x]!= specific_h[x] and specific_h[x]!="0":                    
                    specific_h[x] ="?"
                elif specific_h[x]=="0":
                    specific_h[x]=h[x]
                if h[x]!=general_h[x][x]: #or general_h[x][x]!="?"
                    general_h[x][x] ='?'
            print(specific_h)
            print(general_h)
        if target[i] == "no":    
            flag=0
            for x in range(len(specific_h)): 
                
                if h[x]!= specific_h[x] and flag==0:
                    flag==1
                if general_h[x][x]=="?" or h[x]==general_h[x][x]:
#                     for i in range(0,len(specific_h)):
                    general_h[x][x] = k(h[x],x)
#                     print(general_h)
                check(x)
            print(specific_h)
            print(general_h)
#                 else:                    
#                     general_h[x][x] = '?'
    indices = [i for i, val in enumerate(general_h) if val == ['?', '?', '?', '?', '?']]    
    for i in indices:   
        general_h.remove(['?', '?', '?', '?', '?']) 
    return specific_h, general_h 
s_final, g_final = learn(concepts, target)   
print("Final Specific_h: ", s_final, sep="\n")
print("Final General_h: ", g_final, sep="\n")
```
